## 3. Using the date column, create new columns for: year, month, day of the week and hour of the day.

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day_of_week'] = df['datetime'].dt.dayofweek
df['hour_of_day'] = df['datetime'].dt.hour

df.drop(columns=["datetime"], inplace=True)
```

```python
df = df.reindex(columns=["year", "month", "day_of_week", "hour_of_day"] + list(df.columns[:-4]))
```

## 5. Calculate the total number of casual and registered bikes rented in the years 2011 and 2012.

```python
def calculate_total_bikes_rented(df):
  total_casual_2011 = df[df['year'] == 2011]['casual'].sum()
  total_registered_2011 = df[df['year'] == 2011]['registered'].sum()
    
  total_casual_2012 = df[df['year'] == 2012]['casual'].sum()
  total_registered_2012 = df[df['year'] == 2012]['registered'].sum()

  return total_casual_2011, total_registered_2011, total_casual_2012, total_registered_2012

print(f"Total number of casual  bikes rented in the years 2011:- \033[1;31m{calculate_total_bikes_rented(df)[0]}\033[0m")
print(f"Total number of registered  bikes rented in the years 2011:- \033[1;31m{calculate_total_bikes_rented(df)[1]}\033[0m")
print(" ")
print(f"Total number of casual  bikes rented in the years 2012:- \033[1;31m{calculate_total_bikes_rented(df)[2]}\033[0m")
print(f"Total number of registered  bikes rented in the years 2012:- \033[1;31m{calculate_total_bikes_rented(df)[3]}\033[0m")
```

## 6. Calculate the mean of the hourly total rentals count by season. Which season has the highest mean ?

```python
mean_rentals_by_season = df.groupby('season')['count'].mean()

highest_mean_season = mean_rentals_by_season.idxmax()
highest_mean_value = mean_rentals_by_season.max()

print("Mean of hourly total rentals by season:")
print(mean_rentals_by_season)
print(f"\nSeason with the highest mean hourly rentals: {highest_mean_season} (Mean: {highest_mean_value:.2f})")
```

## 7. Are more bikes rented by registered users on working or non-working days ? Does the answer differ for non-registered users ? Is the answer the same for both years ?

#### Are more bikes rented by registered users on working or non-working days ?

```python
bike_rentals_by_workingday_year = df.groupby(['year', 'workingday'])[['registered']].sum()

print("Total casual and registered bike rentals by year and working day status:")
print(bike_rentals_by_workingday_year)

print("\nAre more bikes rented by registered users on working or non-working days?")
for year in df['year'].unique():
    registered_working = bike_rentals_by_workingday_year.loc[(year, 1), 'registered']
    registered_non_working = bike_rentals_by_workingday_year.loc[(year, 0), 'registered']
    if registered_working > registered_non_working:
        print(f"In {year}, more bikes are rented by registered users on working days ({registered_working}) than non-working days ({registered_non_working}).")
    else:
        print(f"In {year}, more bikes are rented by registered users on non-working days ({registered_non_working}) than working days ({registered_working}).")
```

#### Does the answer differ for non-registered users ?

```python
bike_rentals_by_workingday_year = df.groupby(['year', 'workingday'])[['casual','registered']].sum()
for year in df['year'].unique():
    casual_working = bike_rentals_by_workingday_year.loc[(year, 1), 'casual']
    casual_non_working = bike_rentals_by_workingday_year.loc[(year, 0), 'casual']
    if casual_working > casual_non_working:
        print(f"In {year}, more bikes are rented by non-registered users on working days ({casual_working}) than non-working days ({casual_non_working}).")
    else:
        print(f"In {year}, more bikes are rented by non-registered users on non-working days ({casual_non_working}) than working days ({casual_working}).")
```

####  Is the answer the same for both years ?

```python
registered_working_2011 = bike_rentals_by_workingday_year.loc[(2011, 1), 'registered']
registered_non_working_2011 = bike_rentals_by_workingday_year.loc[(2011, 0), 'registered']
registered_working_2012 = bike_rentals_by_workingday_year.loc[(2012, 1), 'registered']
registered_non_working_2012 = bike_rentals_by_workingday_year.loc[(2012, 0), 'registered']

registered_pattern_2011 = "working days" if registered_working_2011 > registered_non_working_2011 else "non-working days"
registered_pattern_2012 = "working days" if registered_working_2012 > registered_non_working_2012 else "non-working days"

if registered_pattern_2011 == registered_pattern_2012:
    print(f"For registered users, the answer is the same for both years: more bikes are rented on {registered_pattern_2011}.")
else:
    print("For registered users, the answer differs between years.")
```

## 8. Which months in the year 2011 have the highest and the lowest total number of bikes rented ? Repeat for the year 2012.

```python
def get_min_max_rentals_by_month(df, year):
    year_df = df[df['year'] == year]
    monthly_rentals = year_df.groupby('month')['count'].sum()

    highest_month = monthly_rentals.idxmax()
    highest_rentals = monthly_rentals.max()

    lowest_month = monthly_rentals.idxmin()
    lowest_rentals = monthly_rentals.min()

    print(f"\n--- Year {year} ---")
    print(f"Month with highest rentals: {highest_month} (Total: {highest_rentals})")
    print(f"Month with lowest rentals: {lowest_month} (Total: {lowest_rentals})")

get_min_max_rentals_by_month(df, 2011)
get_min_max_rentals_by_month(df, 2012)
```

## 9. Which type of weather have the highest and lowest mean of the hourly total rentals count ?

#### Weather Type -
##### 1: Clear, Few clouds, Partly cloudy, Partly cloudy
##### 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist.
##### 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds.
##### 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog

```python
def get_min_max_averagerentalcount_weather(df, year):
  year_df = df[df['year'] == year]

  average_rentalcount = year_df.groupby('weather')['count'].mean()

  highest_weather = average_rentalcount.idxmax()
  highest_average_rentalcount = average_rentalcount.max()

  lowest_weather = average_rentalcount.idxmin()
  lowest_average_rentalcount = average_rentalcount.min()

  print(f"\n--- Year {year} ---")
  print(f"weather with highest mean of total rental counts: Type - {highest_weather} (Total: {highest_average_rentalcount})")
  print(f"weather with lowest mean of total rental counts: Type - {lowest_weather} (Total: {lowest_average_rentalcount})")

get_min_max_averagerentalcount_weather(df, 2011)
get_min_max_averagerentalcount_weather(df, 2012)
```

### 10. Calculate the correlation between the hourly total rentals count and all the numerical columns in the dataset. Which column has the highest correlation with the total rentals count ?

```python
def highest_correlation_with_count(df):
  numerical_columns = df.drop(columns=['season'])
  correlations = numerical_columns.corr()['count']

  correlations = correlations.drop('count')

  highest_correlation_column = correlations.abs().idxmax()
  highest_correlation_value = correlations[highest_correlation_column]

  print("Correlations with 'count' column:")
  print(correlations)
  print(f"\nColumn with the highest correlation with 'count': {highest_correlation_column} (Correlation: {highest_correlation_value:.4f})")

highest_correlation_with_count(df)
```

### 11. Create a new categorical column called day_period, which can take four possible values: night, morning, afternoon and evening. These values correspond to the following binning of the hour column: 0-6: night, 6-12: morning, 12-6: afternoon, 6-24:evening.

```python
df['day_period'] = pd.cut( df['hour_of_day'], bins=[0, 6, 12, 18, 24], labels=['night', 'morning', 'afternoon', 'evening'], right=False )
```

### 12. Generate a pivot table for the mean of the hourly total rentals count, with the index set to the day period and the column set to the working day column. What can you observe from the table ?

```python
pivot_table = pd.pivot_table(df, values='count', index='day_period', columns='workingday', aggfunc='mean', observed=False)
print(pivot_table)
print(" ")
print("Working days show higher bike rentals during morning and evening commutes. Non-working days see more rentals overnight and throughout the afternoon, suggesting leisure use. Peak usage on non-working days is in the afternoon.")
      
```
